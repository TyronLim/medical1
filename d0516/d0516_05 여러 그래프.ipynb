```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
plt.bar(df['이름'],df['국어'])
```

```python
plt.plot(df['이름'],df['국어'])
```

```python
# subplots 여러개 그래프 생성
fig,axs = plt.subplots(2,2,figsize=(10,10))

# 0,0 번째 그래프
# 범례 : label
axs[0,0].bar(df['이름'],df['국어'],label='국어점수')
axs[0,0].plot(df['이름'],df['영어'],label='영어점수')

# 범례 출력
axs[0,0].legend()

# 상단 타이틀
axs[0,0].set_title('국어 그래프')

# x축 y축 글자
axs[0,0].set(xlabel='이름',ylabel='국어점수')

# 그래프 배경 색상
axs[0,0].set_facecolor('lightyellow')

# 그리드
axs[0,0].grid(ls='--',alpha=0.5)


# 0,1 그래프
axs[0,1].plot(df['이름'],df['영어'],label='영어점수')
axs[0,1].plot(df['이름'],df['수학'],label='수학점수')
axs[0,1].legend(loc='upper center')

# 1,0 그래프
axs[1,0].bar(df['이름'],df['사회'])

# 1,1 그래프, 꺾은 선 그래프

axs[1,1].plot(df['이름'],df['과학'],color='green',alpha=0.5)
```
