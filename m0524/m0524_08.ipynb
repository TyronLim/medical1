```python
df = pd.read_csv('diabetes.csv')
df
```

```python
# df.info()
# df.describe()
df_zero = df
df_zero['Glucose'].replace(0,df_zero['Glucose'].mean(),inplace=True)
df_zero['BloodPressure'].replace(0,df_zero['BloodPressure'].mean(),inplace=True)
df_zero['SkinThickness'].replace(0,df_zero['SkinThickness'].mean(),inplace=True)
df_zero['Insulin'].replace(0,df_zero['Insulin'].mean(),inplace=True)
df_zero['BMI'].replace(0,df_zero['BMI'].mean(),inplace=True)

df_zero
```

```python
input = df_zero.iloc[:,:-1]
target = df_zero['Outcome']


from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(input,target,test_size=0.2,random_state=42)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scale = ss.transform(train_input)
test_scale = ss.transform(test_input)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_scale, train_target)
print(lr.score(train_scale, train_target))
print(lr.score(test_scale, test_target))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
poly.fit(train_scale)
poly.fit(test_scale)
train_poly = poly.transform(train_scale)
test_poly = poly.transform(test_scale)

```

```python
lr_poly = LinearRegression()
lr_poly.fit(train_poly, train_target)
print(lr_poly.score(train_poly, train_target))
print(lr_poly.score(test_poly, test_target))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
rf = RandomForestClassifier(n_jobs=-1, random_state=42)

score = cross_validate(rf,input,target,return_train_score=True,n_jobs=-1)
print(np.mean(score['train_score']),np.mean(score['test_score']))
```
