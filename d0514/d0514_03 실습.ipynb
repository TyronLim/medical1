```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col = '지원번호')
df
```

```python
# 이름, 국어 성적을 가지고 선 그래프를 출력하시오.
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글
# matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'    # 한글
matplotlib.rcParams['font.size'] = 10   # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False     # 마이너스 부호

x = df['이름']
y = df['국어']
plt.title('빅데이터 시작그래프')
plt.xlabel('이름',loc='right')
plt.ylabel('국어점수',loc='top')
plt.bar(x,y,label='국어그래프')
plt.legend(loc='upper center')
```

```python
# chipotle 에서 주문 중 가장 많이 주문한 메뉴 10개를 꺾은 선으로 출력하시오.
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
filt = df.groupby('item_name').count().sort_values('order_id',ascending=False).head(5)
x = filt.index
y = filt['order_id']
x1 = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).index
y1 = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).values
matplotlib.rcParams['font.size'] = 9

plt.figure(dpi=300)
plt.plot(x,y,label='item_count')
plt.plot(x1,y1,label='item_total_count')
plt.legend(loc='upper center')
plt.xlabel('item_name',loc='right',color='red')
plt.ylabel('count',loc='top',color='orange')

```
