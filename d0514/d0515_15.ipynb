```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel('../z20_data/20230104_연령별인구현황2.xlsx',skiprows=3,index_col='행정기관코드')

# df.shape
df
```

```python
df['boy']=0
df['adult']=0
df['old']=0
df.loc[0,'boy']=df.iloc[0,4:6].str.replace(',','').astype(int).sum()
df.loc[0,'adult']=df.iloc[0,6:16].str.replace(',','').astype(int).sum()
df.loc[0,'old']=df.iloc[0,16:24].str.replace(',','').astype(int).sum()

df['boy2']=0
df['adult2']=0
df['old2']=0
df.loc[0,'boy2']=df.iloc[0,49:52].str.replace(',','').astype(int).sum()
df.loc[0,'adult2']=df.iloc[0,52:62].str.replace(',','').astype(int).sum()
df.loc[0,'old2']=df.iloc[0,62:70].str.replace(',','').astype(int).sum()
    
df['boy3']=0
df['adult3']=0
df['old3']=0
df.loc[0,'boy3']=df.iloc[0,95:98].str.replace(',','').astype(int).sum()
df.loc[0,'adult3']=df.iloc[0,98:108].str.replace(',','').astype(int).sum()
df.loc[0,'old3']=df.iloc[0,108:116].str.replace(',','').astype(int).sum()
    
df['boy4']=0
df['adult4']=0
df['old4']=0
for i in range(18):
    df.loc[0,'boy4']=df.iloc[i,141:144].str.replace(',','').astype(int).sum()
for i in range(18):
    df.loc[0,'adult4']=df.iloc[i,144:154].str.replace(',','').astype(int).sum()
for i in range(18):
    df.loc[0,'old4']=df.iloc[i,154:162].str.replace(',','').astype(int).sum()
```

```python
x=np.arange(3)
x = ['2023년 1월','2023년 2월','2023년 3월','2023년 4월']
y3 = df.loc[0,'boy']
y2 = df.loc[0,'adult']
y1 = df.loc[0,'old']

plt.bar(x,y3,width=0.2,label='old')
plt.bar(x,y2,width=0.2,bottom=y3,label='adult')
plt.bar(x,y1,width=0.2,bottom=y2+y3,label='boy')
plt.legend(loc='upper center')
y_1 = [y3,y2,y1]

x_2 = '2024년 2월'
y3_2 = df.loc[0,'boy2']
y2_2 = df.loc[0,'adult2']
y1_2 = df.loc[0,'old2']

plt.bar(x,y3_2,width=0.5,label='old2')
plt.bar(x,y2_2,width=0.5,bottom=y3,label='adult2')
plt.bar(x,y1_2,width=0.5,bottom=y2+y3,label='boy2')
plt.legend(loc='upper center')
y_2 = [y3_2,y2_2,y1_2]

x_3 = '2024년 3월'
y3_3 = df.loc[0,'boy3']
y2_3 = df.loc[0,'adult3']
y1_3 = df.loc[0,'old3']

plt.bar(x,y3_3,width=0.5,label='old3')
plt.bar(x,y2_3,width=0.5,bottom=y3,label='adult3')
plt.bar(x,y1_3,width=0.5,bottom=y2+y3,label='boy3')
plt.legend(loc='upper center')
y_3 = [y3_3,y2_3,y1_3]

x_4 = '2024년 4월'
y3_4 = df.loc[0,'boy4']
y2_4 = df.loc[0,'adult4']
y1_4 = df.loc[0,'old4']

plt.bar(x,y3_4,width=0.5,label='old4')
plt.bar(x,y2_4,width=0.5,bottom=y3,label='adult4')
plt.bar(x,y1_4,width=0.5,bottom=y2+y3,label='boy4')
plt.legend(loc='upper center')
y_4 = [y3_4,y2_4,y1_4]
```
