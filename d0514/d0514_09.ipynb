```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus']=False

import pandas as pd
df = pd.read_csv('../z20_data/stu_score1.csv')
df.head(10)
```

```python
h_df = df.head(20)
x=df.head(20)['id']
y=df.head(20)['total']
y1=df['kor'].head(20)
y2=df['eng'].head(20)
y3=df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(180,300)
plt.xticks(rotation=90)

plt.show()
```
