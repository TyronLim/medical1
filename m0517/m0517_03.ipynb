```python
# random code 0-149 랜덤 숫자 생성

# 랜덤만들기 1

np.random.seed(150)
index = np.arange(150)      # 150개의 배열 생성
np.random.shuffle(index)
index

# 랜덤만들기 2
ran = [n for n in range(149)]
random.shuffle(ran)
ran

# index로 df에 대입해서 가져오면, 랜덤으로 데이터가 섞임!!!
```

```python
df.info()
```

```python
### rows 100개 출력
### 0~99까지의 4개 컬럼을 가진 데이터를 출력하시오.
## 랜덤을 수동으로!!
train_input = df.iloc[index[0:100],0:4]
test_input = df.iloc[index[100:],0:4]
train_target = df.iloc[index[0:100],4]
test_target = df.iloc[index[100:],4]
train_input
test_input
```

```python
## 랜덤을 자동으로 하는 법!! 
# ( 데이터 전처리 : data, result 데이터 분리)
datas_input = df.iloc[:,:-1]
datas_target = df.iloc[:,4]
# 모든 data를 가지로, train, test 데이터 분리
# 자동으로 분리
train_input2,test_input2,train_target2,test_target2 = train_test_split(datas_input,datas_target)
test_input2
```

```python
# 데이터 학습
clf = svm.SVC()
# 데이터 훈련
clf.fit(train_input2,train_target2)

# 데이터 예측하기
result = clf.predict(test_input2)

# 정답률 구하기
# score = metrics.accuracy_score(test_target2,result)           # 정답률 함수1
score = clf.score(test_input2,test_target2)                     # 정답률 함수2
print('정답률 :',score)
```
