```python

category = []
content = []
for i in sites:
    url = i
    browser = webdriver.Chrome()
    try:
        browser.get(url)
        time.sleep(1)
        soup = BeautifulSoup(browser.page_source,'lxml')
        cate = soup.find('li',{'class':'_1bYIAv37'})
        cont = soup.find('div',{'class':'a7c394cI'})
        category.append(cate.text)
        if cont != None:
            content.append(cont.text)
        browser.close()
    except: 
        category.append(0)
        content.append(0)
        print('패스')
```
