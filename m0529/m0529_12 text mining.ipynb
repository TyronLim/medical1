```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('namu.csv',encoding='utf-8-sig',index_col=0)
df.head(2)
```

```python
# 텍스트 전처리
# 정규표현식 (한글이 아닌 것을 찾아서 sub삭제하라)
def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅎ가-힣]')       # [^a] : a가 아닌 것만(not)   //   ^[a]{1} : 첫째자리가 a인가
    result = hangul.sub('',text)
    return result
```

```python
text_cleaning(df['Content'][0])
```

```python
df['Title'] = df['Title'].apply(lambda x: text_cleaning(x))
df['Category'] = df['Category'].apply(lambda x: text_cleaning(x))
df['Content'] = df['Content'].apply(lambda x: text_cleaning(x))
```
