```python
students = []

with open('stu_score1.csv','r',encoding='utf8') as f:
    while True:
        txt = f.readline()
        if txt == '':break
        stu = txt.split(',')
        students.append(stu)

# print(students)    
data_dic = {}
d_list = []
for i,key in enumerate(students):
    for data in students[1:]:
        d_list.append(data[0])
        d_list.append(data[1])
        data_dic[key]=d_list

data_dic
```
