그래프 꾸미기

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글
# matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'    # 한글
matplotlib.rcParams['font.size'] = 15   # 글자 크기
matplotlib.rcParams['axes.unicode_minus']=False     # 마이너스 부호

x=['han','kim','hong']
y=[20,40,80]
plt.plot(x,y)
plt.title('막대그래프',fontdict={'family':'HYGungSo-Bold','size':30})   # 상단 타이틀 글자변경, 크기 변경

# x,y축 문구
plt.xlabel('국어',loc='right',color='red') # 문구, 위치, 색상
plt.ylabel('이름',loc='top',color='green')
# plt.show()
```

```python
x=['han','kim','hong']
y=[20,40,80]
y2=[50,30,90]
y3=[80,90,95]

plt.plot(x,y,label = '국어')
plt.plot(x,y2,label = '영어')
plt.plot(x,y3,label = '수학')

# x,y축 문구
plt.title('막대그래프')
plt.xlabel('국어',loc='right',color='red') # 문구, 위치, 색상
plt.ylabel('이름',loc='top',color='green')
plt.legend(loc='upper left')   # 범례(그래프 설명) , 코드가 있음.
plt.legend(loc=2,ncols=3)   # 범례(그래프 설명) , 코드 사용 , ncols : 범례 1줄에 3개씩 출력 
# plt.cols=3
```
