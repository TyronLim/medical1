```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_excel('../z20_data/score.xlsx')
df.head(10)
```

```python
x = df['이름']
y1 = df['국어']
y2 = df['영어']
y3 = df['수학']
```

```python
plt.bar(x,y1,label='국어')
plt.bar(x,y2,bottom=y1,label='영어')
plt.bar(x,y3,bottom=y1+y2,label='수학')
plt.legend(loc='upper center')
# y=[y1,y2,y3]
# for j in y:
for i,txt in enumerate(y1):
    plt.text(x[i],y1[i]-10,txt,ha='center')
    
for i,txt in enumerate(y2):
    plt.text(x[i],y1[i]+y2[i]-10,txt,ha='center')
    
for i,txt in enumerate(y3):
    plt.text(x[i],y1[i]+y2[i]+y3[i]-10,txt,ha='center')
```
