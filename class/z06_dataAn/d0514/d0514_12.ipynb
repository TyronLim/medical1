```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_csv('../z20_data/drinks.csv')
df
df.fillna('AA',inplace=True)
df
```

```python
# 대륙별 개수를 출력하시오.
labels = df['continent'].value_counts().index
values = df['continent'].value_counts().values
labels
values
```

```python
wedgeprops={'width':0.6, 'edgecolor':'w','linewidth':2}
plt.pie(values,labels=labels,autopct='%.3f%%',wedgeprops=wedgeprops)
plt.title('대륙별 국가 분포도')
explode = [0.1,0,0,0,0,0]
plt.legend(loc=(1.2,0.1),title='<continent>')
plt.show()
```
