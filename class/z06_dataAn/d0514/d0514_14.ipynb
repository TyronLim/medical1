```python
# 막대그래프 score.xlsx 합계 점수를 뽑아서 그래프로 표현하시오.
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus']=False

df = pd.read_csv('../z20_data/score.csv')
df
```

```python
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']

x = df['이름']
y = df['합계']

colors = ['#657d64','#2b402a','#356b32','#2c7528','#42cf3c','#78eb73','#0bff00','#023600']
plt.bar(x,y,label='합계')
plt.title('합계')
plt.grid(axis='y',alpha=0.2,linestyle='--',linewidth=2,color='red')
plt.legend(loc=(1.1,1.1))
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+2,txt,ha='center')
```
