```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

연령별 인구현황 2023 1월 연령별 인구수를 가로막대그래프로 출력하시오.


```python
df = pd.read_excel('../z20_data/20230104_연령별인구현황2.xlsx',skiprows=3,index_col='행정기관코드')
df.head(1)
```

```python
x=df.columns[3:24]
y=(df.iloc[0,3:24].str.replace(',','').astype(int)/1000).values

plt.barh(x,y)
```

```python
x2=df.columns[26:47]
y2=(df.iloc[0,26:47].str.replace(',','').astype(int)/1000).values

plt.barh(x,-y,label='1월남자')
plt.barh(x,y2,label='1월여자')
plt.barh(x2,-y,label='2월남자')
plt.barh(x2,y2,label='2월여자')




plt.legend(loc='upper right')
plt.xlim(-3000,3000)
```
