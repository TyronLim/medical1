```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 설정
matplotlib.rcParams['font.size'] = 10 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('../z20_data/stat_142801.xls',skiprows=2,nrows=2,index_col=0)
df.T
```

```python
# df['2011'].astype(float)
df.info()
```

```python
# 출생아 수는 막대 그래프, 출산율은 꺾은 선 그래프로

plt.figure(figsize=(13,10))
df.loc['출생아\xa0수']
x = df.columns.values
y = df.loc['출생아\xa0수'].values

y2 = (df.loc['합계\xa0출산율'].values)*500
plt.yticks([300,400,500,600])
plt.bar(x,y,color='orange')
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+10,txt,ha='center',color='blue')
plt.plot(x,y2,color='pink',marker='o')
# plt.ylim(0,1400)
for i,txt in enumerate(y2):
    plt.text(x[i],y2[i]+15,txt/500,ha='center',color='black')

```

```python
y3 = (df.loc['합계\xa0출산율'].values)
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.05,txt,ha='center',color='black')
plt.plot(x,y3,color='pink',marker='o',ms=5)
plt.ylim(0,1.5)
plt.yticks([0,1])
```

```python
plt.yticks([300,400,500,600])
plt.ylim(0,700)
plt.bar(x,y,color='orange')
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+10,txt,ha='center',color='blue')
```

```python
fig,ax1 = plt.subplots(figsize=(10,5))
fig.suptitle('출생아수 및 합계출산율')      # 상단 타이틀
ax1.bar(x,y,color='#ff812d')
ax1.set_ylabel('출생아 수')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+3,txt,ha='center',color='black')
    
ax2 = ax1.twinx()
ax2.plot(x,y3,label='합계출산율',color='#ffd100',marker='o',ms=7,lw=1.5,mec='white',mew=2)
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.set_ylabel('합계출산율')
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.03,txt,ha='center',color='black')
```
