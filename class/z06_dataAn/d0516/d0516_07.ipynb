```python
x1 = dfm_1.iloc[0,0:21].str.replace(',','').astype(int).values
y1 = dfw_1.iloc[0,0:21].str.replace(',','').astype(int).values
x2 = dfm_2.iloc[0,0:21].str.replace(',','').astype(int).values
y2 = dfw_2.iloc[0,0:21].str.replace(',','').astype(int).values
x3 = dfm_3.iloc[0,0:21].str.replace(',','').astype(int).values
y3 = dfw_3.iloc[0,0:21].str.replace(',','').astype(int).values
x4 = dfm_4.iloc[0,0:21].str.replace(',','').astype(int).values
y4 = dfw_4.iloc[0,0:21].str.replace(',','').astype(int).values

```

```python
x=[0 for i in range(len(x1))]
for i in range(len(x1)):
    x[i] = (-x1[i]-x2[i]-x3[i]-x4[i])/1000

y=[0 for i in range(len(x1))]
for i in range(len(x1)):
    y[i] = (y1[i]+y2[i]+y3[i]+y4[i])/1000
y
x
```

```python
ind = dfm_1.iloc[0,0:21].index
ind
```

```python
plt.barh(ind,x,label='남자1~4월')
plt.barh(ind,y,label='여자1~4월')
plt.xlim(-10000,10000)
plt.title('남녀 1~4월 인구합(단위:1000명)')
plt.legend(loc='upper right')
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,10))
axs[0,0].barh(ind,-x1,label='1월남자',color='aqua')
axs[0,0].barh(ind,y1,label='1월남자')
axs[0,1].barh(ind,-x2,label='2월남자')
axs[0,1].barh(ind,y2,label='2월남자')
axs[1,0].barh(ind,-x3,label='3월남자')
axs[1,0].barh(ind,y3,label='3월남자')
axs[1,1].barh(ind,-x4,label='4월남자',color='red')
axs[1,1].barh(ind,y4,label='4월남자',color='skyblue')
```
