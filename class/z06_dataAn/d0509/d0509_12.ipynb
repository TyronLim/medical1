```python
import pandas as pd

df = pd.read_csv('../z20_data/stu_score1.csv',index_col='no')
df
```

```python
## total < 220  : 80     total>220   : 60 
# df['science'] =

df['science']=60
filt = df[df['total']<220]

df.loc[filt.index,'science'] = 80
df


# total, avg 부분의 값을 변경

df['total'] = df['kor']+df['eng']+df['math']+df['science']
df['avg'] = df['total']/4
df
```
