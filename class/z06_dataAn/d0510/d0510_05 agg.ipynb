```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score3.csv',index_col=0)
df
```

```python
# 학교별 국어점수가 가장 높은 점수를 출력하시오.
df.groupby('school')['kor'].max()
df.groupby('school')['kor'].mean()
df.groupby(['school','grade'])['kor'].mean()['구로고']
```

```python
# 남자, 여자 수를 출력하시오.
df.groupby('gender').count()
# df.groupby('gender').value_counts()
```

```python
df.describe()
```

```python
# 학교의 남자,여자 국어,영어,수학 평균
df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('kor',ascending=False)
```

```python
df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['디지털고']

# df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('school',ascending=False)
```

```python
df.groupby(['school','gender'])[['kor','eng','math']].mean().iloc[0]
```

```python
# 합계 컬럼을 추가해서 학년별 합계 평균을 출력하시오.
```

```python
df['total'] = 0
df['total'] = df['kor']+df['eng']+df['math']
df.groupby('grade')['total'].mean()
```

## agg   >> 함수 다수 출력
