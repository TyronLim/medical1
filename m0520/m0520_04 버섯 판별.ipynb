```python
# 원 핫 인코딩으로 변환
target = []
data = []
attr_list = []
for row_index,row in df.iterrows():
    target.append(row[0])       # target(결과 데이터) 저장
    row_data = []
    for col,v in enumerate(row.loc[1:]):
        if row_index==0:
            attr = {'dic':{},'cnt':0}   # cnt = 종류 수
            attr_list.append(attr)
        else:
            attr = attr_list[col]
            
        # 원핫인코딩 변환
        d = [0] * 12
        if v in attr['dic']:
            idx = attr['dic'][v]
            print(idx)
        else:
            idx = attr['cnt']
            attr['dic'][v] = idx
            print(attr['dic'][v])
            attr['cnt'] += 1
            
        d[idx] = 1
        row_data += d       # 데이터 저장
            
                    
    data.append(row_data)       # 1줄 리스트를 다시 리스트에 추가

df2 = pd.DataFrame(data)
df2
```
