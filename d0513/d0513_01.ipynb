```python
# 평균 컬럼을 추가해서 grade 컬럼에 A,B,C,D,F 학점을 출력하시오.
df['평균'] = df['합계']/5
```

```python
def avg(avg):
    if avg >=90: return 'A'
    elif avg >=80: return 'B'
    elif avg >=70: return 'C'
    elif avg >=60: return 'D'
    else : return 'F'

df['grade']=0
```

```python
df['grade'] = df['평균'].apply(avg)
df
```

```python
df['SW특기'].fillna('python',inplace=True)
df
```

1. 컬럼을 대문자로 변경하는 방법
 - 결측치를 부여한 후 str함수를 사용해서 대문자로 변경
2. 함수를 사용해서 NaN데이터 처리와 대문자로 변경하는 구문을 사용


```python
df['특기']=df['SW특기'].str.upper()
df
```

```python
def upper_change(txt):
    if pd.isnull(txt): return 'PYTHON'
    else:
        return txt.upper()
```

```python
df['SW특기'] = df['SW특기'].apply(upper_change)
df
```
